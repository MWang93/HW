# Import Data Set

```python
from sklearn import datasets
x, y = datasets.make_moons(200, noise=.05, random_state=0)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x[:,0], x[:,1], c=y, s=50, cmap="viridis")
```

# 1.1 Running K-means

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2, random_state=0).fit(x)
x_predicted = kmeans.predict(x)
```

```python
import numpy as np
sum(np.logical_and(x_predicted, y))/x_predicted.shape[0]
```

```python
def choose_k(k):
    kmeans = KMeans(n_clusters=k, random_state=0).fit(x)
    x_predicted = kmeans.predict(x)
    acc = sum(np.logical_and(x_predicted, y))/x_predicted.shape[0]
    #plt.scatter(x[:,0], x[:,1], c=x_predicted, s=50, cmap="viridis")
    return x_predicted, acc
```
